```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
import matplotlib.pyplot as plt
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)
# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
# Set x limits
plt.xlim(-4.0, 4.0)
# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# Set y limits
plt.ylim(-1.0, 1.0)
# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# Save figure using 72 dots per inch
plt.savefig("exercise_2.png", dpi=72)
```

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')
```

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)
plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.show()
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
plt.show()
```

```python
#Q2
import numpy as np
import matplotlib.pyplot as plt
 
# setting width of bar
barWidth = 0.20
fig = plt.subplots(figsize =(10, 6))
 
# setting height of bar
Result1 = [2, 5, 8, 5]
Result2 = [3, 2, 5, 7]
 
# Setting position of bar on X axis
br1 = np.arange(len(Result1))
br2 = [x + barWidth for x in br1]

# Making the plot
plt.bar(br1, Result1, color ='purple', width = barWidth,
        edgecolor ='black', label ='Result1')
plt.bar(br2, Result2, color ='black', width = barWidth,
        edgecolor ='black', label ='Result2')

# Adding Xticks
plt.xlabel('Method', fontweight ='bold', fontsize = 18)
plt.ylabel('Result Type', fontweight ='bold', fontsize = 18)
plt.xticks([r + barWidth for r in range(len(Result1))],
        ['A', 'B', 'C', 'D'])
 
plt.legend()
plt.show()
```

```python
#Q1--
import numpy as np
import matplotlib.pyplot as plt


a1 = np.pi
x1 = np.arange(-a1, a1, .1)
z1 = np.tan(x1)
plt.plot(x1, z1)
plt.ylim(-10, 10)

a2 = np.pi
x2 = np.arange(-a2, a2, .1)

z2 = 1/np.tan(x2)
plt.plot(x2, z2)
plt.ylim(-10, 10)

a3 = np.pi
x3 = np.arange(-a3, a3, .1)

z3 = 1/np.cos(x3)
plt.plot(x3, z3)
plt.ylim(-10, 10)

a4 = np.pi
x4 = np.arange(-a4, a4, .1)

z4 = 1/np.sin(x4)
plt.plot(x4, z4)
plt.ylim(-10, 10)

plt.plot(x1, z1, color="orange", linewidth=2, linestyle="-", label="tan")
plt.plot(x2, z2, color="red", linewidth=2, linestyle="-", label="cot")
plt.plot(x3, z3, color="black", linewidth=2, linestyle="-", label="sec")
plt.plot(x4, z4, color="purple", linewidth=2, linestyle="-", label="cosec")


plt.legend(loc='upper left')
plt.show()
```
